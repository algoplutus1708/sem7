```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
# --- Add this to Cell [1] ---
# You already have accuracy_score, so add the others
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df = pd.read_csv("7458_emails.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1].values
Y
```

```python
# --- This REPLACES Cell [7] ---
# We will use ONE split for both models
# random_state=42 ensures you get the same split every time you run the code
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
```

```python
# --- This REPLACES Cell [8] ---

print("--- 1. Training Support Vector Machine (SVM) ---")
svc = SVC(C=1.0, kernel='rbf', gamma='auto')

# Fit on the new, single training set
svc.fit(X_train, y_train)

# Predict on the new, single test set
y_pred_svc = svc.predict(X_test)

# Calculate all SVM metrics
acc_svc = accuracy_score(y_test, y_pred_svc)
prec_svc = precision_score(y_test, y_pred_svc)
rec_svc = recall_score(y_test, y_pred_svc)
f1_svc = f1_score(y_test, y_pred_svc)

print("--- SVM Results ---")
print(f"Accuracy:  {acc_svc*100:.2f}%")
print(f"Precision: {prec_svc*100:.2f}%")
print(f"Recall:    {rec_svc*100:.2f}%")
print(f"F1 Score:  {f1_svc*100:.2f}%")
```

```python
# --- This REPLACES Cells [15], [16], [17], and [13] ---

print("\n--- 2. Training K-Nearest Neighbors (KNN) ---")
knn = KNeighborsClassifier(n_neighbors=7)

# Fit on the SAME training set
knn.fit(X_train, y_train)

# Predict on the SAME test set
y_pred_knn = knn.predict(X_test)

# Calculate all KNN metrics
acc_knn = accuracy_score(y_test, y_pred_knn)
prec_knn = precision_score(y_test, y_pred_knn)
rec_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)

print("--- KNN Results ---")
print(f"Accuracy:  {acc_knn*100:.2f}%")
print(f"Precision: {prec_knn*100:.2f}%")
print(f"Recall:    {rec_knn*100:.2f}%")
print(f"F1 Score:  {f1_knn*100:.2f}%")

# --- Add this to your existing KNN cell (Cell [9]) ---
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# (You already have this part from your screenshot)
cm_knn = confusion_matrix(y_test, y_pred_knn)
TN_knn = cm_knn[0, 0]
FP_knn = cm_knn[0, 1]
# FN_knn = ... (You had this in your code)
# TP_knn = ... (You had this in your code)
spec_knn = TN_knn / (TN_knn + FP_knn)

# --- HERE IS THE EXTRA CODE TO ADD ---

# 1. Calculate the other 4 metrics
acc_knn = accuracy_score(y_test, y_pred_knn)
prec_knn = precision_score(y_test, y_pred_knn)
rec_knn = recall_score(y_test, y_pred_knn)
f1_knn = f1_score(y_test, y_pred_knn)

# 2. This is a simpler way to get Error Rate
err_rate_knn = 1 - acc_knn 

# --- Now, print everything together ---
print("--- KNN Model Evaluation ---")
print(f"Accuracy:    {acc_knn*100:.2f}%")
print(f"Precision:   {prec_knn*100:.2f}%")
print(f"Recall:      {rec_knn*100:.2f}%")
print(f"F1 Score:    {f1_knn*100:.2f}%")
print(f"Specificity: {spec_knn*100:.2f}%")
print(f"Error Rate:  {err_rate_knn*100:.2f}%")
```
