# Implement  K-Means  clustering/  hierarchical  clustering  on  sales_data_sample.csv  dataset.  Determine thenumber of clusters using the elbow method.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('7458_sales_data_sample.csv', encoding='unicode_escape')
```

```python
df.info
```

```python
#Columns to Remove
to_drop = ['ADDRESSLINE1', 'ADDRESSLINE2', 'STATE', 'POSTALCODE', 'PHONE']
df = df.drop(to_drop, axis=1)
```

```python
#Check for null values
df.isnull().sum()
```

```python
df.dtypes
```

```python
#ORDERDATE Should be in date time
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
#We need to create some features in order to create cluseters
#Recency: Number of days between customer's latest order and today's date
#Frequency : Number of purchases by the customers
#MonetaryValue : Revenue generated by the customers
import datetime as dt
snapshot_date = df['ORDERDATE'].max() + dt.timedelta(days = 1)
df_RFM = df.groupby(['CUSTOMERNAME']).agg({
    'ORDERDATE' : lambda x : (snapshot_date - x.max()).days,
    'ORDERNUMBER' : 'count',
    'SALES' : 'sum'
})

#Rename the columns
df_RFM.rename(columns = {
    'ORDERDATE' : 'Recency',
    'ORDERNUMBER' : 'Frequency',
    'SALES' : 'MonetaryValue'
}, inplace=True)
```

```python
df_RFM.head()
```

```python
# Divide into segments
# We create 4 quartile ranges
df_RFM['M'] = pd.qcut(df_RFM['MonetaryValue'], q = 4, labels = range(1,5))
df_RFM['R'] = pd.qcut(df_RFM['Recency'], q = 4, labels = list(range(4,0,-1)))
df_RFM['F'] = pd.qcut(df_RFM['Frequency'], q = 4, labels = range(1,5))

df_RFM.head()
```

```python
#Create another column for RFM score
df_RFM['RFM_Score'] = df_RFM[['R', 'M', 'F']].sum(axis=1)
df_RFM.head()
```

# We create levels for our Customers
### RFM Score > 10 : High Value Customers
### RFM Score < 10 and RFM Score >= 6 : Mid Value Customers
### RFM Score < 6 : Low Value Customers


```python
def rfm_level(df):
    if bool(df['RFM_Score'] >= 10):
        return 'High Value Customer'
    
    elif bool(df['RFM_Score'] < 10) and bool(df['RFM_Score'] >= 6):
        return 'Mid Value Customer'
    else:
        return 'Low Value Customer'
df_RFM['RFM_Level'] = df_RFM.apply(rfm_level, axis = 1)
df_RFM.head()
```

```python
# Time to perform KMeans
data = df_RFM[['Recency', 'Frequency', 'MonetaryValue']]
data.head()
```

```python
# Our data is skewed we must remove it by performing log transformation
data_log = np.log(data)
data_log.head()
```

```python
#Standardization 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data_log)
data_normalized = scaler.transform(data_log)
data_normalized = pd.DataFrame(data_normalized, index = data_log.index, columns=data_log.columns)
data_normalized.describe().round(2)
```

```python
#Fit KMeans and use elbow method to choose the number of clusters
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

sse = {}

for k in range(1, 21):
    kmeans = KMeans(n_clusters = k, random_state = 1)
    kmeans.fit(data_normalized)
    sse[k] = kmeans.inertia_
```

```python
plt.figure(figsize=(10,6))
plt.title('The Elbow Method')

plt.xlabel('K')
plt.ylabel('SSE')
plt.style.use('ggplot')

sns.pointplot(x=list(sse.keys()), y = list(sse.values()))
plt.text(4.5, 60, "Largest Angle", bbox = dict(facecolor = 'lightgreen', alpha = 0.5))
plt.show()
```

```python
# --- This REPLACES Cell [20] ---

from sklearn.metrics import silhouette_score

# 1. Fit the final K-Means model with K=5 (as found from Elbow Method)
kmeans = KMeans(n_clusters=5, random_state=1)
kmeans.fit(data_normalized) # Fit the model on the normalized data

# 2. Get the cluster labels
kmeans_labels = kmeans.labels_

# 3. Calculate the Silhouette Score (as required by the manual)
silhouette_avg_kmeans = silhouette_score(data_normalized, kmeans_labels)
print(f"--- K-Means (K=5) Evaluation ---")
print(f"Silhouette Score: {silhouette_avg_kmeans:.4f}")

# 4. Add the cluster labels back to the original RFM dataframe
df_rfm_clustered = df_RFM.copy()
df_rfm_clustered['KMeans_Cluster'] = kmeans_labels

print("\n--- RFM data with K-Means clusters ---")
print(df_rfm_clustered.head())
```

```python
# --- Add this in a NEW CELL at the end ---

print("\n\n--- Method 2: Hierarchical Clustering ---")

import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# 1. Plot the Dendrogram to find optimal clusters 
# We use the same 'data_normalized'
plt.figure(figsize=(15, 8))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Customers (Normalized RFM Data)')
plt.ylabel('Distance (Ward Linkage)')

# 'ward' linkage minimizes the variance of the clusters being merged
dendrogram = sch.dendrogram(sch.linkage(data_normalized, method='ward'))

# Add a horizontal line to suggest a cut
# We can cut at a distance that gives us 5 clusters to compare with K-Means
plt.axhline(y=10, color='r', linestyle='--')
plt.text(0, 10.5, "Suggested cut (gives 5 clusters)", color='r')
plt.show()

# 2. Fit Agglomerative Clustering
# Based on the dendrogram, we'll choose 5 clusters to compare with K-Means
hierarchical_cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
hierarchical_labels = hierarchical_cluster.fit_predict(data_normalized)

# 3. Calculate Silhouette Score for Hierarchical Clustering 
silhouette_avg_hierarchical = silhouette_score(data_normalized, hierarchical_labels)
print(f"\n--- Hierarchical Clustering (K=5) Evaluation ---")
print(f"Silhouette Score: {silhouette_avg_hierarchical:.4f}")

# 4. Add these labels to the dataframe as well
df_rfm_clustered['Hierarchical_Cluster'] = hierarchical_labels

print("\n--- RFM data with Both Cluster types ---")
# Display clusters from both methods side-by-side
print(df_rfm_clustered.head())
```
